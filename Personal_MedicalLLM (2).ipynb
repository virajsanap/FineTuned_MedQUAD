```python
from datasets import load_dataset

dataset = load_dataset("lavita/MedQuAD", split="train")

# Check a sample
print(dataset[0])
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig

model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

quantization_config = BitsAndBytesConfig(load_in_8bit=True)

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    quantization_config=quantization_config
)
```

Base Evaluation

```python
eval_dataset = dataset.select(range(100))  # 100 questions for evaluation

def generate_answer(model, tokenizer, question, max_tokens=100):
    prompt = f"### Instruction:\n{question}\n\n### Response:\n"
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=max_tokens)
    return tokenizer.decode(outputs[0], skip_special_tokens=True).split("### Response:")[-1].strip()

# Generate predictions for the evaluation dataset
base_answers = [generate_answer(model, tokenizer, example["question"]) for example in eval_dataset]
reference_answers = [example["answer"] for example in eval_dataset]
```

```python
import evaluate
import bert_score

# ROUGE
rouge = evaluate.load("rouge")
rouge_scores = rouge.compute(predictions=base_answers, references=reference_answers)

# BLEU
bleu = evaluate.load("bleu")
bleu_scores = bleu.compute(predictions=base_answers, references=[[ref] for ref in reference_answers])

# BERTScore (Semantic accuracy)
P, R, F1 = bert_score.score(base_answers, reference_answers, lang="en")

print("ROUGE-L:", rouge_scores['rougeL'])
print("BLEU:", bleu_scores['bleu'])
print("BERTScore (F1):", F1.mean().item())
```

Foramtting the data

```python
def format_data(example):
    return {
        "text": f"### Instruction:\n{example['question']}\n\n### Response:\n{example['answer']}"
    }

formatted_dataset = dataset.map(format_data)
```

Tonkenizing the data

```python
def tokenize(example):
    tokenized_output = tokenizer(
        example["text"],
        padding="max_length",
        truncation=True,
        max_length=512,
    )
    tokenized_output["labels"] = tokenized_output["input_ids"].copy()
    return tokenized_output

tokenized_dataset = formatted_dataset.map(tokenize, batched=True)
tokenized_dataset.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
```

```python
from google.colab import drive
drive.mount('/content/drive')

# ✅ Define the output directory in Google Drive
output_dir = "/content/drive/MyDrive/tinyllama-finetune-checkpoints"

from peft import LoraConfig, get_peft_model, TaskType
from transformers import TrainingArguments, Trainer

# LoRA configuration
peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.05,
)

# Apply LoRA to the base model
model = get_peft_model(model, peft_config)

# ✅ TrainingArguments using Google Drive path
training_args = TrainingArguments(
    output_dir=output_dir,                      # <- Save checkpoints here
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    num_train_epochs=3,
    learning_rate=2e-4,
    fp16=True,
    logging_steps=50,
    save_total_limit=2,                         # Keep only last 2 checkpoints
    save_steps=200,                             # Save checkpoint every 200 steps
    logging_dir=f"{output_dir}/logs",           # Optional: log dir in same folder
    report_to="none"                            # Prevent unwanted integrations
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
)

# ✅ Automatically resumes if a checkpoint exists
# trainer.train()
```

```python
trainer.train(resume_from_checkpoint=True)
```

More FineTuning

```python
# STEP 1: Install required packages (if not already done)
!pip install transformers peft datasets evaluate bitsandbytes trl

# STEP 2: Imports
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForCausalLM, Trainer, TrainingArguments, EarlyStoppingCallback
from peft import LoraConfig, get_peft_model, TaskType
import torch

# STEP 3: Load MedQuAD dataset and split into train/validation
dataset = load_dataset("lavita/MedQuAD", split="train")
train_dataset = dataset.select(range(40000))
val_dataset = dataset.select(range(40000, 47000))
```

```python
# STEP 4: Prompt engineering function
def format_data(example):
    return {
        "text": f"You are a medical assistant. Answer the following question clearly and accurately.\n\n"
                f"Question: {example['question']}\n\n"
                f"Answer: {example['answer']}"
    }

formatted_train = train_dataset.map(format_data)
formatted_val = val_dataset.map(format_data)

# STEP 5: Load tokenizer
base_model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
tokenizer = AutoTokenizer.from_pretrained(base_model_id)
```

```python
# STEP 6: Tokenize function (adds labels too)
def tokenize(example):
    tokenized = tokenizer(example["text"], padding="max_length", truncation=True, max_length=512)
    tokenized["labels"] = tokenized["input_ids"].copy()
    return tokenized

tokenized_train = formatted_train.map(tokenize, batched=True)
tokenized_val = formatted_val.map(tokenize, batched=True)

tokenized_train.set_format("torch", columns=["input_ids", "attention_mask", "labels"])
tokenized_val.set_format("torch", columns=["input_ids", "attention_mask", "labels"])

# STEP 7: Load base model
from transformers import BitsAndBytesConfig

bnb_config = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_threshold=6.0,
    llm_int8_skip_modules=None
)

model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    quantization_config=bnb_config,
    device_map="auto"
)

# STEP 8: Apply LoRA
peft_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,
    lora_alpha=32,
    lora_dropout=0.05,
)
model = get_peft_model(model, peft_config)
```

```python
# STEP 9: Save to Google Drive
from google.colab import drive
drive.mount('/content/drive')
output_dir = "/content/drive/MyDrive/tinyllama-finetune-checkpoints-v2"

# STEP 10: Define training arguments (4 epochs + early stopping)
training_args = TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    num_train_epochs=4,
    learning_rate=2e-4,
    fp16=True,
    evaluation_strategy="steps",
    eval_steps=200,
    save_steps=200,
    save_total_limit=2,
    logging_steps=50,
    load_best_model_at_end=True,
    metric_for_best_model="eval_loss",
    greater_is_better=False,
    report_to="none"
)

# STEP 11: Define Trainer with early stopping
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
)

# STEP 12: Start training
trainer.train()
```
